```python
# lọc ra 100 chuyến đi đầu tiên (theo trip_id) từ gps_trips_part1.csv và lưu vào file mới
import pandas as pd
import os
df = pd.read_csv('../../../Data/Bronze_layer/enriched_weather_data_part1.csv')
# Lấy 100 trip_id đầu tiên
first_100_trip_ids = df['trip_id'].unique()[:100]

df_first_trips = df[df['trip_id'].isin(first_100_trip_ids)]
df_first_trips.to_csv('gps_trips_first5.csv', index=False)
print(f'Đã lưu {len(df_first_trips)} dòng từ {len(first_100_trip_ids)} chuyến đi đầu tiên')
```
